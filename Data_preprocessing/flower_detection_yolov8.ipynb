```python
from ultralytics import YOLO
```

```python
model = YOLO("/Users/choismn/Downloads/best (1).pt")
```

```python
import os
path = []
li = os.listdir("/Users/choismn/Downloads/도라지꽃")
for l in li:
    path.append(os.path.join("/Users/choismn/Downloads/도라지꽃", l))
```

```python
result = model.predict(path)
```

```python
result[0][0]
```

```python
#1번클래스 flower, 2번클래스 Flower 이 있는거만 걸렀음
real_result = []
for i in range(0, len(path)):
    if result[i] and len(result[i]) > 0:
        if hasattr(result[i][0], 'boxes') and hasattr(result[i][0].boxes, 'cls'):
            if 1.0 in result[i][0].boxes.cls or 2.0 in result[i][0].boxes.cls:
                real_result.append([result[i][0].boxes.xywh, result[i][0].path])
```
